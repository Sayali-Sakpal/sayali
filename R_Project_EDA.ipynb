```r
install.packages("dplyr")
library(dplyr)
```

```r
install.packages("readr")
library("readr")
```

```r
install.packages("readxl")
library(readxl)
```

```r
gfg_data=read_excel("/FlipkartProduct.xlsx")
head(gfg_data)
```

```r
str(gfg_data)
```

```r
summary(gfg_data)
```

```r
library(dplyr)
select(gfg_data,id,title,Rating,maincateg,star_5f,star_4f,star_3f,star_2f,star_1f)
head(gfg_data)
```

```r
install.packages("ggplot2")
library("ggplot2")
ggplot(data = gfg_data, aes(x = Rating, y = star_5f))
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_point()
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_density_2d() +
  geom_point()
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_bin2d() +
  geom_point()
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_point(alpha = 0.3, color = "green", size = 3)+ geom_density_2d(color = "red")
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_point(alpha = 0.5) +
  geom_smooth(se=FALSE)
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f, color = star_5f)) +
  geom_point()+geom_smooth(se=FALSE)
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f, color = star_4f)) +
  geom_point()
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_1f)) +
  geom_point() +
  geom_smooth(aes(color = Rating))
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_5f)) +
  geom_point() +
  geom_smooth(aes(color = Rating),se=FALSE)
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_2f)) +
  geom_point() +
  geom_smooth(aes(color = Rating)) +
  facet_wrap( ~ Rating)
```

```r
ggplot(data = gfg_data, mapping= aes(x = Rating, y = star_2f)) +
  geom_point() +
  geom_smooth(aes(color = Rating)) +
  facet_wrap(~ maincateg)
```

```r
ggplot(data = gfg_data, aes(x = Rating)) +
  geom_histogram(bins = 5)
```
